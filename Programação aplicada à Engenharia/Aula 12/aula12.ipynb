## Visualização de dados com Matplotlib

A visualização de dados simplifica a compreensão de conjuntos complexos por meio de gráficos ou tabelas, facilitando a tomada de decisões eficientes e a identificação de tendências. Essa prática é essencial em análises avançadas, como aprendizado de máquina e exploração de dados.

### Matplotlib

[Matplotlib](https://matplotlib.org/) é uma biblioteca abrangente para criar visualizações estáticas, animadas e interativas em Python. 


### Introdução ao pyplot

**matplotlib.pyplot** é uma coleção de funções. [Pyplot](https://matplotlib.org/stable/tutorials/pyplot.html), um módulo do Matplotlib, possibilita realizar modificações em figuras de forma eficiente, utilizando Python. Essa ferramenta é gratuita e de código aberto. Cada função do pyplot desempenha diversas ações em uma figura, como criar a figura, estabelecer uma área de plotagem, desenhar linhas, adicionar rótulos e muito mais. Suporta a criação de diversos tipos de gráficos, incluindo histogramas, dispersões, gráficos 3D, imagens, contornos e polares.

```python
# Lendo a base de dados tips.csv
data = pd.read_csv('tips.csv')

# Inicializando os dados para visualização
x = data['day']
y = data['total_bill']

data
```

Agora iremos criar uma representação de um gráfico simples de barras através do Matplotlib e pyplot.

### Bar plot

Um **gráfico de barras** é um gráfico contendo barras retangulares com comprimentos e alturas proporcionais aos valores que representam. 

Podem ser plotados horizontalmente ou verticalmente, descrevendo comparações entre as categorias discretas (conjunto finito ou enumerável de valores possíveis).

```python
# Plotando os dados
plt.bar(x, y)

# Adicionando título à figura
plt.title("Tips Dataset")
 
# Adicionando a legenda no eixo y
plt.ylabel('Total da Conta')
 
# Adicionando a legenda no eixo x
plt.xlabel('Dia')
 
plt.show()
```

O Matplotlib automaticamente limita os valores e os marcadores (pontos) dos eixos X e Y, mas podemos colocar um limite e os marcadores de maneira manual com as funções xlim() e ylim() para os eixos X e Y, e as funções xticks() e yticks() para os rótulos.

Exemplo: Neste exemplo, alteraremos o limite do eixo Y e definiremos os rótulos do eixo X.

Além disso, temos algumas customizações para o gráfico de barras:
     - **cor**, **edgecolor** (bordas da barra), **linewidth** (largura das bordas da barra) e **width** (largura da barra)

```python
# plotting the data
plt.bar(x, y, color='gray', edgecolor='blue', 
        linewidth=2)

# Adicionando título à figura
plt.title("Tips Dataset", fontsize=18, color="blue")
 
# Adicionando a legenda no eixo y
plt.ylabel('Total da Conta')

# Limitando o eixo y
plt.ylim(0, 100) 

# Adicionando a legenda no eixo x
plt.xlabel('Dia')

plt.show()
```

```python
g = data['tip']

dia_semana = ['Quinta','Sexta','Sábado','Domingo']
media_gorjeta = np.zeros((4))
media_gorjeta[0] = np.mean(g[x=='Quinta'])
media_gorjeta[1] = np.mean(g[x=='Sexta'])
media_gorjeta[2] = np.mean(g[x=='Sábado'])
media_gorjeta[3] = np.mean(g[x=='Domingo'])
```

```python
# Plotando os dados
plt.plot(dia_semana, media_gorjeta, color='orange', linewidth=3, marker='o', 
         markersize=15, linestyle='--')

# Adicionando título à figura
plt.title("Tips Dataset", fontsize=18, color="green")

# Adicionando a legenda no eixo y
plt.ylabel('Média de gorjetas por dia')
 
# Adicionando a legenda no eixo x
plt.xlabel('Dia da semana')

# Colocando marcadores manualmente no Eixo x
plt.xticks(dia_semana, labels=["Domingo", "Sábado", "Quinta-feira", "Sexta-feira"])
 
plt.show()
```

### Histograma

Um histograma é uma representação gráfica de dados que utiliza barras de alturas distintas. Cada barra em um histograma agrupa números em intervalos, e a altura das barras indica a quantidade de dados presente em cada intervalo. 

Barras mais altas indicam uma maior concentração de dados naquele intervalo específico. Essa visualização é útil para observar a forma e a distribuição de dados amostrais contínuos.

Personalizações disponíveis para o histograma incluem **bins**, **color**, **edgecolor**, **linestyle** e **alpha** (0 para transparente e 1 para opaco).

```python
# initializing the data
x = data['total_bill']

# plotting the data
plt.hist(x, bins=25, color='pink', edgecolor='blue',
         linestyle='--', alpha=0.3)
 
# Adding title to the plot
plt.title("Tips Dataset", color='blue')
 
# Adicionando a legenda no eixo y
plt.ylabel('Frequência',color='blue')
 
# Adicionando a legenda no eixo x
plt.xlabel('Total pago pelos clientes', color='blue')
 
plt.show()
```

### Scatter Plot

Os diagramas de dispersão, também conhecidos como gráficos de dispersão, são representações visuais de dados que envolvem duas ou mais variáveis, organizadas em um gráfico.

Utilizando coordenadas cartesianas, esses gráficos exibem os valores de um conjunto de dados, proporcionando uma relação visual entre as variáveis.

Personalizações disponíveis incluem: s (marker size), c (color), marker, linewidths, edgecolor, alpha:

```python
# Lendo o arquivo tips.csv
x = data['size']
y = data['total_bill']
 
# plotando os dados
plt.scatter(x, y, c=data['tip'], s=data['total_bill'], marker='s', alpha=0.5)
 
# Adding title to the plot
plt.title("Tips Dataset")
 
# Adicionando a legenda no eixo y
plt.ylabel('Total pago')
 
# Adicionando a legenda no eixo x
plt.xlabel('Quantidade de pessoas')
 
plt.show()
```

### Pie Chart

O gráfico de setores, também conhecido como gráfico circular ou tradicionalmente chamado de gráfico de pizza, é um diagrama em forma de círculo no qual os valores de cada categoria estatística são representados proporcionalmente às medidas angulares correspondentes.

Por padrão, o gráfico é desenhado iniciando as fatias em 0 graus. Podemos alterar a posição inicial usando o parâmetro **startangle**.

```python
# Lendo o arquivo tips.csv
data = pd.read_csv('tips.csv')

plt.figure(figsize=(8,8))

# plotando os dados
plt.pie(media_gorjeta, labels=["Domingo", "Sábado", "Quinta-feira", "Sexta-feira"], 
        colors=['#a3e4d7', '#aed6f1', '#f9e79f', '#edbb99'], startangle=100, 
        textprops={'fontsize':12}, autopct='%1.2f%%') # autopct: mostrar porcentagem com duas casas decimais
 
# Adicionando título ao enredo
plt.title("Média do valor pago  por dia")
 
plt.show()
```

Outras personalizações para o gráfico de pizza: **explode** (move as fatias do gráfico), **shadow** (cria sombra na borda).

```python
# Lendo o arquivo tips.csv
data = pd.read_csv('tips.csv')

plt.figure(figsize=(6,6))

explode = [0.0, 0.0, 0.0, 0.3]

# plotando os dados
plt.pie(media_gorjeta, labels=["Domingo", "Sábado", "Quinta-feira", "Sexta-feira"], shadow=True,
        colors=['#a3e4d7', '#aed6f1', '#f9e79f', '#edbb99'], startangle=50, explode=explode,
        textprops={'fontsize':12}, autopct='%1.2f%%') # autopct: mostrar porcentagem com duas casas decimais

# Adicionando título ao enredo
plt.title("Média do valor pago  por dia")

plt.show()
```
